# 1. Find the total number of patients of each disease (diagnosis) and plot in histogram.


```python
import matplotlib as plt
from matplotlib import pyplot
```

```python
plt.pyplot.title("number of patient of each disease")
plt.pyplot.xlabel("number of pateint")
plt.pyplot.ylabel("disease")
plt.pyplot.hist(df['diagnosis'].value_counts(),bins=50)
```

# 2. Find the total number of patients of each disease (diagnosis) with respect to the disease state(malignant or benign) and plot in a pie chart.

```python
df['diagnosis'].value_counts()
```

```python
plt.pyplot.pie(df.groupby('diagnosis')['state'].value_counts(),autopct='%1.1f%%')
plt.rcParams["figure.figsize"]= (12,34)
plt.pyplot.title("number of pateint of each disease with respect to status",color='purple',fontsize=19)
```

# 3. Find and plot via pia chart the total number of patients of each disease (diagnosis) with respect to the gender(Female, Male) 

```python
plt.rcParams["figure.figsize"]= (12,34)
plt.pyplot.pie(df.groupby('diagnosis')['gender'].value_counts(),autopct='%1.1f%%')
plt.pyplot.title("number of pateint of each disease with respect to gender",color='blue',fontsize=19)
```
